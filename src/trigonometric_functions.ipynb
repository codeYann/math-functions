```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting $f(x) = \sin(x)$ function

```python
space = np.linspace(-2*np.pi, 2*np.pi)
```

```python
plt.grid()
plt.plot(space, np.sin(space))
plt.show()
```

## 

## Plotting $f(x) = \cos(x)$ function

```python
plt.grid()
plt.plot(space, np.cos(space))
plt.show()
```

## Plotting a Circumference

```python
circle_space = np.linspace(0, 2 * np.pi, 100)

cos_space = np.cos(circle_space)
sin_space = np.sin(circle_space)

fig, ax = plt.subplots()

ax.grid()
ax.plot(cos_space, sin_space)
ax.set_aspect(1)

plt.show()
```

## Plotting $f(x) = \tan(x)$ function

```python
plt.grid()
plt.plot(space, np.tan(space))
plt.show()
```
